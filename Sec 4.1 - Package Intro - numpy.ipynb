## Generate a image by RGB color channels
***

```python
import matplotlib.pyplot as plt

# Image Size (height x width)
h,w = 200,255  # Wider image to show smooth transition
# Create gradient matrices
R = np.linspace(255,100,w,dtype = np.uint8).reshape(1,-1) * np.ones((h,1),dtype = np.uint8)
G = np.linspace(155,255,w,dtype = np.uint8).reshape(1,-1) * np.ones((h,1),dtype = np.uint8)
B = np.linspace(100,255,w,dtype = np.uint8).reshape(1,-1) * np.ones((h,1),dtype = np.uint8)
# Stack the three channels to form an RGB image
RGB_image = np.stack((R,G,B),axis = -1)
# Display the image
plt.imshow(RGB_image)
```

## Slowness of Loops
***
##### ns, nanoseconds = $10^{-9}$ s
##### $\mu \text{s}$, microseconds = $10^{-6}$ s
##### ms, milliseconds = $10^{-3}$ s

```python
def cal_recip(seq):
    res = np.empty(len(seq))
    for k in range(len(res)):
        res[k] = 1 / seq[k]
    return res
vals = np.random.randint(1,7,100000)
%timeit cal_recip(vals)
%timeit 1 / vals
```

## Vectorized function

```python
def heaviside_1(x):
    return 1 if x > 0 else 0
display(heaviside_1(-1),
        heaviside_1(1.5))
x = np.linspace(-1,1,11)
try:
    heaviside_1(x)
except Exception as e:
    print(e)
finally:
    # np.vectorize works with arrays, it will be relatively slow
    heaviside_2 = np.vectorize(heaviside_1)
    display(heaviside_2(x))
```

##### $M_{(2,3)}$ and $A_{(3,)}$  
##### By rule 1, $M_{(2,3)}$ and $A_{(1,3)}$
##### By rule 2, $M_{(2,3)}$ and $A_{(2,3)}$
##### Match

## Case 2
##### $M_{(3,1)}$ and $A_{(3,)}$  
##### By rule 1, $M_{(3,1)}$ and $A_{(1,3)}$
##### By rule 2, $M_{(3,3)}$ and $A_{(3,3)}$
##### Match

```python
M = np.array([[0],[1],[2]])
a = np.array([0,1,2])
M + a
```

```python
M = np.ones((3,2))
a = np.arange(3)
try:
	M + a
except Exception as e:
	print(e)
```

## Gauss-Siedel

```python
# Matrix - based formula
def Gauss_Siedel(A,b,x,tol = 1e-3):
	L = np.tril(A)
	U = A - L
	count = 0
	print('init x: {:.4f} {:.4f} {:.4f}'.format(*x.ravel()))
	while True:
		xc = np.linalg.inv(L) @ (b - (U @ x))
		count += 1
		if np.sqrt((xc - x).T @ (xc - x)) < tol:
			print('it converged \n')
			break
		rpt_str = f'iter {count}'
		rpt_str += ': {:.4f} {:.4f} {:.4f}'.format(*xc.ravel())
		print(rpt_str)
		x = xc
	return x
```

```python
A = np.array([[ 8, 3,-3],
			  [-2,-8, 5],
			  [ 3, 5,10]])
b = np.array([[14],
			   [5],
			  [-8]])
x = np.zeros((3,1))
res_conv = Gauss_Siedel(A,b,x)
print('converged result is \n (x1,x2,x3) = ({:.4f},{:.4f},{:.4f})'.format(*res_conv.ravel()))
```

## Projection

```python
X = np.array([1.47,1.5,1.52,1.55,1.57,
              1.6,1.63,1.65,1.68,1.7,
              1.73,1.75,1.78,1.80,1.83])
v1 = np.ones(X.shape[0])
# model matrix
Xm = np.vstack([v1,X]).T
y = np.array([52.21,53.12,54.48,55.84,57.2,
              58.57,59.93,61.29,63.11,64.47,
              66.28,68.1,69.92,72.19,74.46]).reshape(-1,1)
display(Xm,y)
```

```python
H1 = Xm @ np.linalg.inv(Xm.T @ Xm) @ Xm.T
```

```python
Q,R = np.linalg.qr(Xm)
H2 = Q @ Q.T
```

```python
np.hstack([H1 @ y,H2 @ y])  # 兩種方法結果相同
```
